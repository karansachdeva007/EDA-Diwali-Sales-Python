```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sales=pd.read_csv('Diwali Sales Data.csv',encoding="ISO-8859-1")
sales
# encoding="unicode_escape" --> to avoid error
```

```python
sales.drop(['Status','unnamed1'],axis=1,inplace=True)
# drop unrelated/blank columns
```

```python
sales.shape
```

```python
sales.isnull().sum()
# check for null values
```

```python
sales.dropna(inplace=True)
# drop nulll values
```

```python
# change data type
sales['Amount']=sales['Amount'].astype('int')
```

```python
sales['Amount'].dtype
```

```python
sales.rename(columns={'Marital_Status':'Shaadi'})
```

# EXPLORATORY DATA ANALYSIS

```python
sales.columns
```

```python
ax=sns.countplot(x="Gender",data=sales)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_gen=sales.groupby('Gender',as_index=False)['Amount'].sum().sort_values(ascending=False,by='Amount')
sales_gen
```

```python
sales_gen=sales.groupby('Gender',as_index=False)['Amount'].sum().sort_values(ascending=False,by='Amount')
sns.barplot(x="Gender",y="Amount",data=sales_gen)
plt.show()
```

from above 2 graphs we can observe that females are the most buyers and have more purchasing power

```python
# Age group with gender

ax=sns.countplot(x='Age Group',data=sales,hue="Gender",palette="prism_r")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
# age group with Amount

sales_ag=sales.groupby('Age Group',as_index=False)['Amount'].sum().sort_values(ascending=False,by='Amount')
sns.barplot(x="Age Group",y="Amount",data=sales_ag)
plt.show()
```

From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

```python
# total number of orders from top 15 states
sales_st=sales.groupby('State',as_index=False)['Orders'].sum().sort_values(ascending=False,by='Orders').head(15)
sns.barplot(x="State",y="Orders",data=sales_st)
sns.set(rc={'figure.figsize':(15,5)})
plt.xticks(rotation=75)
plt.show()
```

```python
# total amount from top 15 states
sales_am=sales.groupby('State',as_index=False)['Amount'].sum().sort_values(ascending=False,by='Amount').head(15)
sns.barplot(x="State",y="Amount",data=sales_am,color="Green")
sns.set(rc={'figure.figsize':(15,5)})
plt.xticks(rotation=75)
plt.show()
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

# Marital Status

```python
ax=sns.countplot(x='Marital_Status',data=sales,hue="Marital_Status",palette="prism")
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_state = sales.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender',palette="rainbow")
plt.show()
```

from above graphs we can see that most of the buyers are married (women) and they have high purchasing power

# Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = sales, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_state = sales.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
plt.show()
```

 From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

# Product category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = sales, x = 'Product_Category',hue="Product_Category")
plt.xticks(rotation=75)

plt.show()
```

```python
sales_state = sales.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount',hue='Product_Category',palette="viridis")
plt.show()
```

 From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
sales_state = sales.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
plt.show()
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
sales.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
plt.show()
```

Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
